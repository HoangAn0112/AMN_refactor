On little dataset the QÂ² drastically depends on the train test splitting !

```python
import tensorflow as tf
from amn.model import AMNWtModel
from sklearn.ensemble import GradientBoostingRegressor
from amn.tools import MaxScaler
from sklearn.metrics import r2_score

data_dir = "../data"

dataset_file = "/Dataset/iML1515_EXP_UB.npz"
objective=['BIOMASS_Ec_iML1515_core_75p37M']

# dataset_file = "/Dataset/iML1515_EXP_2_UB.npz"
# objective=['BIOMASS_Ec_iML1515_core_75p37M']


scores = []

for i in range(100):
    seed = 10 + i
    tf.random.set_seed(seed)

    model = AMNWtModel(dataset_file=data_dir + dataset_file, 
                       objective=objective,
                       timestep=4,
                       hidden_dim=50,
                       verbose=True,
                       uptake_max_index = None)

    # Preprocessing
    scaler= MaxScaler()
    model.train_test_split(test_size=0.1, random_state=seed)
    model.preprocess(scaler)
    model.preprocessing_for_specific_model()

    gb_reg = GradientBoostingRegressor()
    gb_reg.fit(model.X_train, model.Y_train[:,0])

    R_2 = r2_score(gb_reg.predict(model.X_train), model.Y_train[:,0])
    Q_2 = r2_score(gb_reg.predict(model.X_test), model.Y_test[:,0])

    scores.append([R_2,Q_2])
```
